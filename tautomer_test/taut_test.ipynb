```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
```

```python
from molvs import tautomer
from molvs import standardize
from molvs import standardize_smiles
```

```python
t1 = Chem.MolFromSmiles('C1=CC=CNC(=O)1')
t2 = Chem.MolFromSmiles( 'c1ccc(O)nc1' )
```

```python
Draw.MolsToGridImage([t1,t2])
```

```python
enum=tautomer.TautomerEnumerator()
```

```python
ts=enum.enumerate(t1)
ts2 = enum.enumerate(t2)
```

```python
Draw.MolsToGridImage(ts)
```

```python
Draw.MolsToGridImage(ts2)
```

```python
canon = tautomer.TautomerCanonicalizer()
```

```python
canon.canonicalize(t1)
```

```python
canon.canonicalize(t2)
```
